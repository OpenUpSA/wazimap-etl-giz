```python
import pandas as pd

df = pd.read_csv (r'V2_KfW.csv')
df
```

```python
df['Programme'].unique()
```

```python
input_column_name = "Programme"


unique_values = df[input_column_name].unique()
for value in unique_values:
    df1 = df.loc[df[input_column_name] == value]
    filename = 'V2_'+value +'.csv'
    print(filename)
    df1.to_csv(filename, index = False)
```

```python
for service in df['Service'].unique():
    count = df.loc[df['Service'] == service].count()
    filename = service +'.csv'
    print(filename)
    print(count) 
```
